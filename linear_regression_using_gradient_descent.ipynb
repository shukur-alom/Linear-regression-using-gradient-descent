```python
import pandas as pd
import numpy as np
import math
import random
```

```python
df = pd.read_csv("/content/edit.csv")
```

```python
df.head(5)
```

```python
x = df.drop('pre',axis=1).to_numpy()
y = df[['pre']].to_numpy()
m = df.shape[1] 
```

```python
def cost_function(h_x_y_squer,m):
  return (1*h_x_y_squer)/(2*m)

def weight_update(h_x_y,m,p_w,alpha):
  return (p_w-((alpha*h_x_y)/m))
```

```python
w = [] 
w_0 = 0
for  i in range(x.shape[1]):w.append(0.0)
w = np.array(w)
loss_data = []
```

```python
epoch = 1075756756009
call_back = 1000
alpha = 0.001
for i in range(epoch):
  h_x_y = 0
  h_x_y_squer = 0
  for j in range(m):
    d = x[j]*w
    h_x = 0 
    for k in d: h_x += k
    h_x += w_0
    h_x_y += h_x - y[j][0]
    h_x_y_squer += math.pow((h_x - y[j][0]), 2)
    loss_f = cost_function(h_x_y_squer,m)
    loss_data.append(loss_f)
  if loss_f >= 1.0:
    for l in range(w.shape[0]):
      w[l] = float(weight_update(h_x_y,m,w[l],alpha))
    w = np.array(w)
    w_0 = weight_update(h_x_y,m,w_0,alpha) 
  elif loss_f >= 0.0:
    print("\n\n\n\t\t\tBreak Epoch for loss is 0")
    break
  if loss_data.count(loss_f) == call_back:
    print("\n\n\n\t\t\tBreak Epoch for callback")
    break
  print(f"Epoch : {i}; Loss : {loss_f}; Weights : {w} ; C : {w_0}")
```

```python
#test data
for i in range(df.shape[0]):
  result = 0
  ff = x[i]*w
  for q in ff:
    result += q
  result += w_0
  print(f"Predict : {result}; True Value : {y[i][0]}")
```
